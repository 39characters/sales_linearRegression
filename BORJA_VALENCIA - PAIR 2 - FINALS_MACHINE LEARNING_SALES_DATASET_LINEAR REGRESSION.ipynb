```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt

df = pd.read_csv('sales_data.csv')
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns

X = df['Price']  # Independent variable
y = df['Quantity Sold']  # Dependent variable

X = sm.add_constant(X)

model = sm.OLS(y, X).fit()

print(model.summary())

y_pred = model.predict(X)

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Price', y='Quantity Sold', data=df, label='Data')
plt.plot(df['Price'], y_pred, color='red', label='Regression Line')
plt.xlabel('Price')
plt.ylabel('Quantity Sold')
plt.title('Linear Regression: Price vs. Quantity Sold')
plt.legend()
plt.show()
```
